```python
import numpy as np
import pandas as pd
import sys
sys.path.append("../fraud_detection/src/")

from util import s_to_time_format, string_to_datetime,hour_to_range
from tqdm import tqdm

#-----------------------------
# load data
#-----------------------------
df_train = pd.read_csv("/data/yunrui_li/fraud/dataset/train.csv")
df_test = pd.read_csv("/data/yunrui_li/fraud/dataset/test.csv")


for df in [df_train, df_test]:
    # pre-processing
    df["loctm_"] = df.loctm.astype(int).astype(str)
    df.loctm_ = df.loctm_.apply(s_to_time_format).apply(string_to_datetime)
    # time-related feature
    df["loctm_hour_of_day"] = df.loctm_.apply(lambda x: x.hour)
    df["loctm_minute_of_hour"] = df.loctm_.apply(lambda x: x.minute)
    df["loctm_second_of_min"] = df.loctm_.apply(lambda x: x.second)

    # removed the columns no need
    df.drop(columns = ["loctm_", "loctm","txkey"], axis = 1, inplace = True)
```

```python
def num_transaction_in_past_n_days(df, n):
    """
    Calculate how many transaction that this user have in the past n days
    """
    current_day_at_this_transaction = df.locdt.tolist()
    conam_at_this_transaction = df.conam.tolist()
    #print (current_day_at_this_transaction)
    #print (conam_at_this_transaction)
    output = []
    conam_out = []
    for current_date, current_conam in zip(current_day_at_this_transaction,conam_at_this_transaction):
        history_date = current_date-n
        c = 0
        for i in current_day_at_this_transaction:
            if (i >= history_date) & (i < current_date):
                c+=1
                conam_out.append(current_conam)
        output.append(c)
#     print ("sum",sum(conam_out))
#     print ("min",pd.Series(conam_out).min())
#     print ("max",pd.Series(conam_out).max())
#     print ("mean",pd.Series(conam_out).mean())

    return sum(conam_out) # return Series instead of list

def stats_in_past_n_days(df, n, method = "sum"):
    """
    Calculate stats that a bacno have in the past n days. Noted df should be sorted already.
    """
    current_day_at_this_transaction = df.locdt.tolist()
    conam_at_this_transaction = df.conam.tolist()
    #print (current_day_at_this_transaction)
    #print (conam_at_this_transaction)
    output = []
    conam_out = []
    for current_date, current_conam in zip(current_day_at_this_transaction,conam_at_this_transaction):
        history_date = current_date-n
        c = 0
        for i in current_day_at_this_transaction:
            if (i >= history_date) & (i < current_date):
                c+=1
                conam_out.append(current_conam)
        output.append(c)
    if method == "sum":
        return sum(conam_out) # return Series instead of list
    elif method == "mean":
        return pd.Series(conam_out).mean()
    elif method == "max":
        return pd.Series(conam_out).max()
    elif method == "min":
        return pd.Series(conam_out).min()
    else:
        raise ValueError("this method {} we don't support".format(method))
        
# df = pd.concat([df_train, df_test], axis = 0)
# df.sort_values(by = ["bacno","locdt"], inplace = True)

# df = df.head(1000)
# for past_n_days in [3,7,14,30]:
#     for m in [
#         #"sum",
#           "mean",
#           #"max",
#           #"min"
#     ]:
#         cols = "agg_{}_{}_conam_in_past_{}_days".format("bacno",m,past_n_days)
#         print (cols)
#         tmp = df[["bacno","locdt","conam"]].groupby("bacno")\
#             .apply(lambda x: stats_in_past_n_days(x,past_n_days,m)).to_frame(cols).reset_index()
#         df = df.merge(tmp, on = "bacno", how = "left")

# df = pd.concat([df_train, df_test], axis = 0)
# df.sort_values(by = ["mchno","locdt"], inplace = True)

# df = df.head(1000)
# for past_n_days in [3,7,14,30]:
#     for m in [
#           "sum",
#           "mean",
#           "max",
#           "min",
#     ]:
#         cols = "agg_{}_{}_conam_in_past_{}_days".format("mchno",m,past_n_days)
#         print (cols)
#         tmp = df[["mchno","locdt","conam"]].groupby("mchno")\
#             .apply(lambda x: stats_in_past_n_days(x,past_n_days,m)).to_frame(cols).reset_index()
#         df = df.merge(tmp, on = "mchno", how = "left")

df = pd.concat([df_train, df_test], axis = 0)
df.sort_values(by = ["hcefg","locdt"], inplace = True)

#df = df.head(100000)
for past_n_days in [3,7,14,30]:
    for m in [
          #"sum",
          "mean",
          #"max",
          #"min",
    ]:
        cols = "agg_{}_{}_conam_in_past_{}_days".format("hcefg",m,past_n_days)
        print (cols)
        tmp = df[["hcefg","locdt","conam"]].groupby("hcefg")\
            .apply(lambda x: stats_in_past_n_days(x,past_n_days,m)).to_frame(cols).reset_index()
        df = df.merge(tmp, on = "hcefg", how = "left")
```

```python
def filter_(df,s,e):
    out = []
    print 
    for ix, row in df.iterrows():
        if row.locdt <= e and row.locdt >=s:
            out.append(row.conam)
    return out
```

```python
def average_stats_in_past_n_days(df, n, method):
    """
    
    """
    time_interval = [(i+1,i+n) for i in range(120-n+1)]

    out = []
    for s,e in time_interval:
        tmp = df[(df.locdt <=e) & (df.locdt >= s)]
        #print (len(tmp))
        if len(tmp!=0):
            if method == "mean":
                out.append(tmp.conam.mean())
            elif method == "sum":
                out.append(tmp.conam.sum())
            elif method == "max":
                out.append(tmp.conam.max())
            elif method == "min":
                out.append(tmp.conam.min())
            elif method == "std":
                out.append(tmp.conam.std())
            else:
                raise ValueError("this method {} we don't support".format(method))

    return 1.0* sum(out)/len(out)

def average_stats_in_past_n_days(df, n, method):
    """
    
    """
    time_interval = [(i+1,i+n) for i in range(120-n+1)]

    out = []
    for s,e in time_interval:
        
        tmp = filter_(df,s,e)
        #print (len(tmp))
        if len(tmp!=0):
            if method == "mean":
                out.append(pd.Series(tmp).mean())
            elif method == "sum":
                out.append(pd.Series(tmp).sum())
            elif method == "max":
                out.append(pd.Series(tmp).max())
            elif method == "min":
                out.append(pd.Series(tmp).min())
            elif method == "std":
                out.append(pd.Series(tmp).std())
            else:
                raise ValueError("this method {} we don't support".format(method))

    return 1.0* sum(out)/len(out)

def get_agg_time_feature(df, gby = "hcefg"):
    for past_n_days in [3,7,14,30]:
        for m in ["mean","sum","max","min","std"]:
            cols = "agg_{}_{}_conam_in_past_{}_days".format(gby,m,past_n_days)
            tmp = df[["{}".format(gby),"locdt","conam"]].groupby(gby).apply(lambda x: average_stats_in_past_n_days(x,past_n_days,m)).to_frame(cols).reset_index()
            df = df.merge(tmp, on = gby, how = "left")
        break
    return df   

import time
df = pd.concat([df_train, df_test], axis = 0)
df.sort_values(by = ["hcefg","locdt"], inplace = True)
df = df.head(1000)
s = time.time()
df = get_agg_time_feature(df,gby = "mcc")
e = time.time()
print (e-s)
# for past_n_days in [3,7,14,30]:
#     for m in ["mean","sum","max","min","std"]:
#         cols = "agg_{}_{}_conam_in_past_{}_days".format("hcefg",m,past_n_days)
#         tmp = df[["hcefg","locdt","conam"]].groupby("hcefg").apply(lambda x: average_stats_in_past_n_days(x,past_n_days,m)).to_frame(cols).reset_index()
#         df = df.merge(tmp, on = "hcefg", how = "left")
#     break
```
