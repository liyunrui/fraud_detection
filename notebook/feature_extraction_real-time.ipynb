```python
import numpy as np
import pandas as pd
import sys
sys.path.append("../fraud_detection/src/")

from util import s_to_time_format, string_to_datetime,hour_to_range
from tqdm import tqdm

#-----------------------------
# load data
#-----------------------------
df_train = pd.read_csv("/data/yunrui_li/fraud/dataset/train.csv")
df_test = pd.read_csv("/data/yunrui_li/fraud/dataset/test.csv")


for df in [df_train, df_test]:
    # pre-processing
    df["loctm_"] = df.loctm.astype(int).astype(str)
    df.loctm_ = df.loctm_.apply(s_to_time_format).apply(string_to_datetime)
    # time-related feature
    df["loctm_hour_of_day"] = df.loctm_.apply(lambda x: x.hour)
    df["loctm_minute_of_hour"] = df.loctm_.apply(lambda x: x.minute)
    df["loctm_second_of_min"] = df.loctm_.apply(lambda x: x.second)

    # removed the columns no need
    df.drop(columns = ["loctm_", "loctm","txkey"], axis = 1, inplace = True)
```

```python
from tqdm import tqdm
for gby in [
    #"scity",
#     "stocn",
#     "acqic",
#     "mchno",
    "bacno",
    "cano"
#     "mcc",
#     "hcefg",
#     "contp",
#     "etymd"
]:
    df = pd.concat([df_train, df_test], axis = 0)
    df.sort_values(by = [gby,"locdt"], inplace = True)
    df = df.head(1000)
    feat = [gby,"locdt","conam"]
    df = df[feat]
    for n in [
              7,
              14]:
        for method in [
            #"min",
            #"max",
            "mean",
            #"std",
            #"median"
        ]:
            cols = "{}_{}_conam_in_past_{}_days".format(gby,method,n)
            ans = []
            key1 = []
            key2 = []
            for _,x in tqdm(df[feat].groupby([gby,"locdt"])):
                s,e = _[1]-n,_[1]
                tmp = df[df[gby] ==_[0]]
                tmp = tmp[(tmp.locdt < e) & (tmp.locdt >= s)]
                if len(tmp)!=0:
                    if method == "mean":
                        ans.append(tmp.conam.mean())
                    elif method == "sum":
                        ans.append(tmp.conam.sum())
                    elif method == "max":
                        ans.append(tmp.conam.max())
                    elif method == "min":
                        ans.append(tmp.conam.min())
                    elif method == "std":
                        ans.append(tmp.conam.std())
                    elif method == "median":
                        ans.append(tmp.conam.median())  
                else:
                    ans.append(0.0)
                key1.append(_[0])
                key2.append(_[1])
            tmp_ = pd.DataFrame({gby:key1,"locdt":key2,cols:ans})
            tmp_.to_csv("../fraud_detection/features/{}.csv".format(cols), index = False)
            print (cols)

            break
```
